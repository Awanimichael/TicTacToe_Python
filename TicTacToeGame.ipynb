```python
import random

class BoardGame:
    
    #init to initialise list of board for a TicTaoToe game
    def __init__(self):
        self.board = [" "] * 9
        
    def display(self):
        print(self.board)
        
    def makeMove(self):
    #using list comprehension to create choices that computer can pick from
        size    = len(self.board)
        choices = [i for i in range(size) if self.board[i] == " "]
        if len(choices) == 0:
            return -1
        myChoice = random.choice(choices)
        return myChoice
    
    def humanMove(self):
        choices = [j for j in range(len(self.board)) if self.board[j] == " "]
        cell = int(input("Enter an index number (0-9) for your move :"))
        if cell in choices:
            return cell
        else:
            print("Invalid Move: This spot is taken by the other player or does not exit on the board")
            return self.humanMove()
        
    
    def humanBoardUpdate(self, cell, player):
        if self.board[cell] != " ":
            return "Invalid Cell"
        else:
            self.board[cell] = player
            return ""
            
    def computerBoardUpdate(self, cell, player):
        cell = self.makeMove()
        if cell == -1:
            return "D"
        self.board[cell] = player
        return ""

            
    def isThereAWinner(self, player):
        winningMoves = [[0,3,6],[1,4,7],[2,5,8],
                        [0,4,8],[2,4,6],[0,1,2],
                        [3,4,5],[6,7,8]]
        winString = ""

        for i, j, k in winningMoves:
            winString = self.board[i] + self.board[j] + self.board[k]
            if winString   == "OOO":
                return True
            elif winString == "XXX":
                return True
        return False
    
    
def playGame():
    ticTacToe = BoardGame()
    outcome = ""
    
    while not outcome:
        ticTacToe.display()
        
        #Get human move and Update the Board
        human = ticTacToe.humanMove()
        outcome = ticTacToe.humanBoardUpdate(human, "X")
        if outcome == "Invalid Cell":
            print("That is an invalid move")
        ticTacToe.display()
        
        #Check if its a winner
        if ticTacToe.isThereAWinner("X"):
            outcome = "X"
            print("Gave Over!, Human wins")
        else:
            #get computer move and Update the Board
            computer = ticTacToe.makeMove()
            outcome = ticTacToe.computerBoardUpdate(computer, "O")
            
            if outcome == "D":
                print("Game Over!, Its a Tie")
            
            #check if its a winner
            if ticTacToe.isThereAWinner("O"):
                outcome = "O"
                print("Gave Over!, Computer wins")
            else:
                continue
                
if __name__ == "__main__": 
    playGame()
```
